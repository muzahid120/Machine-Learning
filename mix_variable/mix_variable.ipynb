```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')# cpture numeric part 
df['cabin_cat']=df['Cabin'].str[0] # capture first lettr 
```

```python
df['Ticket_num']=df['Ticket'].apply(lambda s:s.split()[-1])# extract the last bit of ticket as a number 

df['Ticket_num']=pd.to_numeric(df['Ticket_num'],errors='coerce',downcast='integer')# convert str to number 
```

```python
df['Ticket_cat']=df['Ticket'].apply(lambda s:s.split()[0])## extract the first part of ticket as category
df['Ticket_cat']=np.where(df['Ticket_cat'].str.isdigit(),np.nan,df['Ticket_cat'])# where number come replace that nan 
```

```python
df['cabin_num']=pd.to_numeric(df['cabin_num'],errors='coerce',downcast='integer')# convert str to number
```

```python
df.drop(columns=['Ticket','Cabin'],inplace=True)
```
