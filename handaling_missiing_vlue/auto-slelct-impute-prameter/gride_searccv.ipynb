```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv(r'C:\Users\SDS\Downloads\train.csv')
df=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
df
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
numeric_feature=[2,5]
catagorical_feature=[1,6]
```

```python
num_transform=Pipeline(steps=[('imputer',SimpleImputer(strategy='median')),
('scaled',StandardScaler())])
```

```python
cat_transform=Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
('onehot',OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessing=ColumnTransformer(transformers=[('num',num_transform,numeric_feature),
('cat',cat_transform,catagorical_feature)])
```

```python
clf=Pipeline([('preprossor',preprocessing),
('model',LogisticRegression())])
```

```python
from sklearn import set_config

set_config(display='diagram')
clf
```

```python
param_grid = {
    'preprocessor__imputer__strategy': ['mean', 'median', 'most_frequent'],
    'model__C': [0.1, 1.0, 10.0],
    'model__solver': ['liblinear', 'lbfgs']
}

```

```python
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)

# Print the best parameters
print(f"Best params:")
print(grid_search.best_params_)
```
